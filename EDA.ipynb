```python
# Definir una función de recomendación de juegos
def recomendacion_juego(item_id, num_recomendaciones=5):
    # Obtener las características del juego seleccionado por item_id
    juego_seleccionado = features[features['item_id'] == item_id]
    
    if juego_seleccionado.empty:
        raise ValueError("El item_id no se encuentra en los datos")
    
    # Calcular la similitud del coseno entre el juego seleccionado y todos los juegos
    cosine_sim = cosine_similarity(juego_seleccionado.iloc[:, :-2], features.iloc[:, :-2])
    
    # Obtener los índices de los juegos más similares
    sim_scores = list(enumerate(cosine_sim[0]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_similar_games = sim_scores[1:(num_recomendaciones+1)]
    
    # Obtener los nombres de los juegos recomendados
    recomendaciones = [features.iloc[i[0]]['item_name'] for i in top_similar_games]
    
    return recomendaciones
```

```python
# Función de recomendación
def recomendacion_juego(item_id: int):
    # Obtener el índice del juego basado en su 'item_id'
    idx = data.index[data['item_id'] == item_id].tolist()
    
    if not idx:
        # Si el índice está vacío, significa que el item_id no se encontró en los datos
        raise IndexError("El item_id no se encuentra en los datos")
    
    idx = idx[0]
    
    # Calcular la similitud del juego con otros juegos
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Ordenar los juegos por similitud
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Obtener las 5 recomendaciones más similares (excluyendo el juego de entrada)
    top_similar_games = sim_scores[1:6]
    
    # Obtener los nombres de los juegos recomendados
    result = [data['item_name'][i[0]] for i in top_similar_games]

    return result
```
